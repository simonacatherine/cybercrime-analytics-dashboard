```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cybercrime_9A1.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df = df.dropna()
df.info()
```

```python
df['Chargesheet_Rate_2024'].head(10)
```

```python
df['Chargesheet_Rate_2024'] = pd.to_numeric(
    df['Chargesheet_Rate_2024'],
    errors='coerce'
)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.replace("-", np.nan, inplace=True)
```

```python
df.drop(columns=['SL'], inplace=True)
```

```python
df["Growth_22_23"] = (
    (df["Cases_2023"] - df["Cases_2022"])
    / df["Cases_2022"]
) * 100
```

```python
df.head()
```

```python
df["Growth_23_24"] = (
    (df["Cases_2024"] - df["Cases_2023"])
    / df["Cases_2023"]
) * 100
```

```python
top_states = df.sort_values(
    by="Cases_2024",
    ascending=False
)

print(top_states[
    ["State_UT", "Cases_2024"]
].head(10))
```

```python
top10 = df.nlargest(10, "Cases_2024")

plt.figure(figsize=(10,6))
plt.bar(top10["State_UT"],
        top10["Cases_2024"])

plt.xticks(rotation=45)
plt.title("Top 10 States by Cybercrime Cases (2024)")
plt.xlabel("State")
plt.ylabel("Cases")
plt.show()
```

```python
df.to_csv(
    "cybercrime_9A1_cleaned.csv",
    index=False
)
```

```python
growth = df.sort_values(
    by="Growth_23_24",
    ascending=False
)

print(growth[
    ["State_UT", "Growth_23_24"]
].head(10))
```

```python
crime_rate = df.sort_values(
    by="Crime_Rate_2024",
    ascending=False
)

print(crime_rate[
    ["State_UT", "Crime_Rate_2024"]
].head(10))
```

```python
top10 = df.nlargest(10, "Cases_2024")

plt.figure(figsize=(12,6))

for state in top10["State_UT"]:
    row = df[df["State_UT"] == state]

    plt.plot(
        [2022, 2023, 2024],
        [
            row["Cases_2022"].values[0],
            row["Cases_2023"].values[0],
            row["Cases_2024"].values[0]
        ],
        marker='o',
        label=state
    )

plt.legend()
plt.title("Cybercrime Growth Trends")
plt.xlabel("Year")
plt.ylabel("Cases")

plt.show()
```

```python
decline = df[df["Growth_23_24"] < 0]

print(
    decline[
        ["State_UT", "Growth_23_24"]
    ].sort_values(
        by="Growth_23_24"
    )
)
```

```python
correlation = df[
    ["Population_Lakhs", "Cases_2024"]
].corr()

print(correlation)
plt.figure(figsize=(8,6))

plt.scatter(
    df["Population_Lakhs"],
    df["Cases_2024"]
)

plt.xlabel("Population (Lakhs)")
plt.ylabel("Cybercrime Cases 2024")
plt.title("Population vs Cybercrime Cases")

plt.show()
```

```python
top10 = df.nlargest(10, "Cases_2024")

plt.figure(figsize=(10,6))

plt.scatter(
    df["Cases_2024"],
    df["Crime_Rate_2024"]
)

for i in range(len(top10)):
    plt.annotate(
        top10.iloc[i]["State_UT"],
        (
            top10.iloc[i]["Cases_2024"],
            top10.iloc[i]["Crime_Rate_2024"]
        )
    )

plt.xlabel("Cybercrime Cases (2024)")
plt.ylabel("Crime Rate (2024)")
plt.title("Cybercrime Cases vs Crime Rate")

plt.show()
```
